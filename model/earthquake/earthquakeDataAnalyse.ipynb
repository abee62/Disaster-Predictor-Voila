```python

import folium

def plot_density_mapbox():
    # Create a map centered on the specified coordinates
    out = widgets.Output()

    m = folium.Map(location=[0, 200], zoom_start=2)
    
    # Add a marker for each data point
    for _, row in tsunamiDf.iterrows():
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=6,
            fill=True,
            fill_color='red',
            fill_opacity=0.7,
            popup=f"Magnitude: {row['magnitude']}",
        ).add_to(m)
    display(m)
    # Display the map
    

# Call the function to plot the density map
plot_density_mapbox()
```
